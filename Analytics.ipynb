```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import mpl_toolkits
from mpl_toolkits.mplot3d import Axes3D


def fn(x, y):
    return (4. - 2.1*(x**2) + (x**4)/3)*(x**2) + x*y + (4*(y**2) - 4)*(y**2)

x = np.linspace(-5, 5, 150)
y = np.linspace(-5, 5, 150)

X, Y = np.meshgrid(x, y)

Z = fn(X,Y)

fig = plt.figure(figsize=(15, 15))


spec = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)
for i, max_temp in enumerate([1, 2, 10, 7500]):
    ax = fig.add_subplot(spec[i])
    sns.heatmap(Z, vmax=max_temp, ax=ax)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

data = np.genfromtxt('data/data.csv', delimiter=',')

fig = plt.figure(figsize=(15, 7))
spec = gridspec.GridSpec(ncols=2, nrows=1, figure=fig)

ax = fig.add_subplot(spec[0])
for i in range(2):
    ax.plot(data[:, i], label=f'x{i+1}')
    ax.text(data.shape[0]-100, data[-1, i]-0.02, f'{data[-1, i]:.4f}')

plt.xlabel("Optimization step")
plt.ylabel("Value")
plt.grid(linestyle = '--', linewidth = 0.5)
plt.legend()

ax = fig.add_subplot(spec[1])
ax.plot(data[:, 2], label=f'f')
ax.text(data.shape[0]-100, data[-1, 2]+1000, f'{data[-1, 2]:.4f}')
plt.xlabel("Optimization step")
plt.ylabel("Function Value")
plt.grid(linestyle = '--', linewidth = 0.5)

plt.savefig('data/res.jpg')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec


datasets = ["HDAX", "FTSE", "HSCI", "SP500"]

fig = plt.figure(figsize=(20, 20))
spec = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)

for i, dataset in enumerate(datasets):
    data = pd.read_csv(f'data/wk_price_{dataset}.txt', sep="   ", header=None, engine='python')
    ax = fig.add_subplot(spec[i])
    ax.set_title(dataset)
    for asset in data:
        ax.plot(data[asset])

%reset -f
```
